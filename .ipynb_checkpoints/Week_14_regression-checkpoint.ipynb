```python
#use AdaBoostRegressor to find best features to use
X = diabetes_df.drop('Outcome',axis=1)
y = diabetes_df['Outcome']

#choose 2 features to use later
estimator = AdaBoostRegressor(random_state=0, n_estimators=100)
selector = RFE(estimator, n_features_to_select=2, step=1)
selector = selector.fit(X, y)
 
filter = selector.support_
ranking = selector.ranking_

print("Mask data: ", filter)
print("Ranking: ", ranking)
```

```python
#now to show what features were selected
features = np.array(X.columns)
print("All features:")
print(features)

print("Selected features:")
print(features[filter])
```

```python
#first ran with no interaction
diabetes_vs_glucose_BMI_no_int = logit('Outcome ~ Glucose + BMI', data=diabetes_df).fit()
intercept,ic_glucose,slope = diabetes_vs_glucose_BMI_no_int.params
print(diabetes_vs_glucose_BMI_no_int.params)
```

```python
sns.scatterplot(x='Glucose',
               y='Outcome',
                hue='BMI',
               data=diabetes_df)
sns.set(rc = {'figure.figsize':(8,4)})
plt.title('Outcome Vs Glucose/BMI', fontsize=14)
plt.axline(xy1=(0,intercept),slope=slope,color="black")
plt.show()
```

outcome = -7.52 + .036*Glucose + .08*BMI 

```python
#then ran with interaction

diabetes_vs_glucose_BMI_int = logit('Outcome ~ Glucose * BMI', data=diabetes_df).fit()
intercept,ic_glucose,ic_bmi,slope = diabetes_vs_glucose_BMI_int.params
print(diabetes_vs_glucose_BMI_int.params)
```

```python
# Set Glucose to sequence of 0 to 200 in steps of 10 (based on min and max in dataframe)
Glucose = np.arange(0,200,10)
# Set BMI to sequence of 0 to 70 in steps of 10 (based on min and max in dataframe)
BMI = np.arange(0,70,10)
# Create p as all combinations of values of Glucose and BMI
p = product(Glucose,BMI)
# Transform p to a DataFrame and name the columns
explanatory_data = pd.DataFrame(p, columns = ['Glucose','BMI'])
```

```python
# Create prediction data
prediction_data = explanatory_data.assign(has_diabetes = diabetes_vs_glucose_BMI_int.predict(explanatory_data))
# Create most_likely_outcome
prediction_data['most_likely_outcome'] = np.round(prediction_data['has_diabetes'])
```

```python
plt.title('Outcome Vs Glucose/BMI', fontsize=14)
sns.scatterplot(x='Glucose', y='BMI',data=diabetes_df, hue='Outcome')
sns.scatterplot(x='Glucose', y='BMI',data=prediction_data, hue='most_likely_outcome',alpha=0.2, legend=False)
plt.axline(xy1=(0,intercept),slope=ic_glucose,color="black")
plt.show()

```

```python
print(prediction_data)
```

outcome = -10.32 + .08*Glucose + .16*BMI - 0.0007*Glucose*BMI

```python
conf_matrix = diabetes_vs_glucose_BMI_int.pred_table()
print(conf_matrix)
```

```python
# Extract TN, TP, FN, and FP from conf_matrix
TN = conf_matrix[0,0]
TP = conf_matrix[1,1]
FN = conf_matrix[1,0]
FP = conf_matrix[0,1]

# Calculate and print the accuracy
accuracy = (TN + TP) / (TN + FN + FP + TP)
print('accuracy', accuracy)

# Calculate and print the sensitivity
sensitivity = TP / (TP + FN)
print('sensitivity', sensitivity)

# Calculate and print the specificity
specificity = TN / (TN + FP)
print('specificity', specificity)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(conf_matrix)
ax.grid(False)
plt.title('Confusion Matrix', fontsize=14)
ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j, i, conf_matrix[i, j], ha='center', va='center', color='white')
plt.show()
```

### 6.	What are the other ways to assess the performance of the model besides accuracy? What approach makes the most sense with the diabetes dataset? Why? 

Mean Absolute Error (MAE) - measured by taking the average of the absolute difference between actual values and the predictions.

Root Mean Square Error (RMSE) - measured by taking the square root of the average of the squared difference between the prediction and the actual value. It represents the sample standard deviation of the differences between predicted values and observed values(also called residuals).

Coefficient of determination (R2) - measures how well the actual outcomes are replicated by the regression line.


```python
# find r2/accuracy score
perc = model.score(X_test, y_test)
#round to two decimals
perc_val = round(perc,2)
print(f'R2 Score :{perc_val}')

#find mse
mse = mean_squared_error(y_test, y_pred)
#round to two decimals
mse_val = round(mse,2)
print(f'Mean Squared Error :{mse_val}')

#find rmse
rmse = np.sqrt(mse)
#round to two decimals
rmse_val = round(rmse,2)
print(f'Root Mean Squared Error:{rmse_val}')
```

### 8.	Is the KNN model or the multiple logistic regression model more performant for the diabetes dataset? How can you tell?

After tuning the multiple logistic regression model, the mse is .24 and R2 is .76. Using KNN, the mse is .25 and R2 is .75. The smaller mse and higher R2 belongs to the multiple logistic regression model so it is more performant for the diabetes dataset.

```python
# set up the X and y from class
X = diabetes_df.drop('Outcome',axis=1)
y = diabetes_df['Outcome']

#set same split as done for logistic regression
X_train_knn, X_test_knn, y_train_knn, y_test_knn = train_test_split(X, y, test_size=0.4, random_state=42)

#Standardization - transforming values so mean is 0 and SD is 1
sc = StandardScaler()

X_train_knn = sc.fit_transform(X_train_knn)
X_test_knn = sc.fit_transform(X_test_knn)

#using number of neighbors from class exercise a few weeks ago
knn=KNeighborsClassifier(n_neighbors=25)
knn.fit(X_train_knn, y_train_knn)
y_predict_knn = knn.predict(X_test_knn)
knn.score(X_test_knn,y_test_knn)
```

```python
mse_knn = mean_squared_error(y_test_knn, y_predict_knn)
#round to two decimals
mse_val_knn = round(mse_knn,2)
print(f'Mean Squared Error :{mse_val_knn}')

#find rmse
rmse_knn = np.sqrt(mse_knn)
#round to two decimals
rmse_val_knn = round(rmse_knn,2)
print(f'Root Mean Squared Error:{rmse_val_knn}')
```
